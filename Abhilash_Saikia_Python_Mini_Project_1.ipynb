```python
print(" Welcome to Library Management System ")
library = {"The Palace of illusions": {"author": "Chitra Banarjee Divakaruni", "available": True, "issued_to": None},
    "Death,an inside story.": {"author": "Sadhguru", "available": False, "issued_to": "Abhilash"},
    "Origin": {"author": "Dan Brown", "available": True, "issued_to": None}}
credentials = { "admin": "admin123",
    "librarian": "lib123"}
def login():
    attempts = 3
    while attempts > 0:
        username = input("Enter username: ")
        password = input("Enter password: ")

        if username in credentials and credentials[username] == password:
            print("Login successful!")
            return True
        else:
            attempts -= 1
            print(f"Invalid credentials. Attempts left: {attempts}")

    print("Too many failed attempts. Exiting system.")
    return False
def display_books():
    if not library:
        print("No books available in the library.")
        return
    print("Available Books:")
    print("----------------------------")
    for title, details in library.items():
        status = "Available" if details["available"] else f"Issued to {details['issued_to']}"
        print(f"Title: {title}")
        print(f"Author: {details['author']}")
        print(f"Status: {status}")
        print("----------------------------")
def add_book():
    title = input("Enter book title: ")
    author = input("Enter author name: ")
    if title in library:
        print("Book already exists in the library.")
    else:
        library[title] = {"author": author,"available": True,"issued_to": None}
        print("Book added successfully!")
def issue_book():
    title = input("Enter book title to issue: ")
    if title in library:
        if library[title]["available"]:
            name = input("Enter name of the person issuing the book: ")
            library[title]["available"] = False
            library[title]["issued_to"] = name
            print("Book issued successfully!\n")
        else:
            print("Book is already issued.")
    else:
        print("Book not found in the library.")
def return_book():
    title = input("Enter book title to return: ")
    if title in library:
        if not library[title]["available"]:
            library[title]["available"] = True
            library[title]["issued_to"] = None
            print("Book returned successfully!")
        else:
            print("Book is already available.")
    else:
        print("Book not found in the library.")
def view_issued_books():
    issued = False
    print("\nIssued Books:")
    print("----------------------------")
    for title, details in library.items():
        if not details["available"]:
            print(f"Title: {title}")
            print(f"Issued To: {details['issued_to']}")
            print("----------------------------")
            issued = True
    if not issued:
        print("No books are currently issued.")
def main_menu():
    while True:
        print("\nLibrary Menu")
        print("1. View Books")
        print("2. Add a Book")
        print("3. Issue a Book")
        print("4. Return a Book")
        print("5. View Issued Books")
        print("6. Exit")
        choice = input("Enter your choice (1-6): ")
        if choice == "1":
            display_books()
        elif choice == "2":
            add_book()
        elif choice == "3":
            issue_book()
        elif choice == "4":
            return_book()
        elif choice == "5":
            view_issued_books()
        elif choice == "6":
            print("Thank you for using the Library Management System!")
            break
        else:
            print("Invalid choice. Please try again.")
if login():
    main_menu()
```
